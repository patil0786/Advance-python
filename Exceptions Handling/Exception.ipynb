# Generic exception handling


```python
def division2():

    try:

        num = int(input("Please enter Numerator"))
        den = int(input("Please enter Denominator"))
        div = num/den
        print(f"Numerator :{num},Denominator : {den},Division : {div}")
        return div
    except Exception as e:
        print(f"Exception occured : {e}")
```

```python

d2 = division2()
```

```python
d2 = division2()
```

```python
d2 = division2()
```

# Errors handleling for different exception
1. zero division error
2. Value error

```python
def division3():

    try:

        num = int(input("Please enter Numerator"))
        den = int(input("Please enter Denominator"))
        div = num/den
        print(f"Numerator : {num},Denominator :{den},Division : {div}")
        return div
    except ZeroDivisionError:
        print("Denominator in division can not be zero")
    except ValueError:
        print("Numerator and Denominator should be both number ,Blank input can not be allowed")
```

```python
d3 = division3()
```

```python
d3 = division3()
```
